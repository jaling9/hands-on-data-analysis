### 2.4 数据的合并

#### 2.4.1 任务一：将data文件夹里面的所有数据都载入，观察数据的之间的关系

```python
#写入代码
train_left_down = pd.read_csv('data/train-left-down.csv')
train_left_up.head()
```

```python
#写入代码
train_left_up = pd.read_csv('data/train-left-up.csv')
train_left_up.head()
```

```python
train_right_down = pd.read_csv('data/train-right-down.csv')
train_left_down.head()
```

```python
train_right_up = pd.read_csv('data/train-right-up.csv')
train_right_up.head()
```

【提示】结合之前我们加载的train.csv数据，大致预测一下上面的数据是什么

#### 2.4.2：任务二：使用concat方法：将数据train-left-up.csv和train-right-up.csv横向合并为一张表，并保存这张表为result_up

```python
#写入代码
result_up = pd.concat([train_left_up,train_right_up],axis=1) # 注意连接的是表名，不需要加引号
result_up.head(2)
```

#### 2.4.3 任务三：使用concat方法：将train-left-down和train-right-down横向合并为一张表，并保存这张表为result_down。然后将上边的result_up和result_down纵向合并为result。

```python
#写入代码
result_down = pd.concat([train_left_down,train_right_down],axis=1) # 注意连接的是表名，不需要加引号
result = pd.concat([result_up,result_down],axis=0) # 这里需要上下拼接
result.head()
# pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, 
# levels=None, names=None, verify_integrity=False, sort=False, copy=True)[source]
```

#### 2.4.4 任务四：使用DataFrame自带的方法join方法和append：完成任务二和任务三的任务

```python
#写入代码
result_down_for_join = train_left_down.join(train_right_down)
result_up_for_join = train_left_up.join(train_right_up)
result_for_join = result_up_for_join.append(result_down_for_join)
result_for_join 
# append垂直方向连接，join水平方向连接
# DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)
# DataFrame.append(other, ignore_index=False, verify_integrity=False, sort=False)
```

#### 2.4.5 任务五：使用Panads的merge方法和DataFrame的append方法：完成任务二和任务三的任务

```python
#写入代码
result_down_for_merge = pd.merge(train_left_down,train_right_down,left_index=True, right_index=True)
result_up_for_merge = pd.merge(train_left_up,train_right_up,left_index=True, right_index=True)
result_for_merge = result_up_for_join.append(result_down_for_join)
result_for_merge
# pandas.merge(left, right, how='inner', on=None, left_on=None, right_on=None, 
    # left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)
```

【思考】对比merge、join以及concat的方法的不同以及相同。思考一下在任务四和任务五的情况下，为什么都要求使用DataFrame的append方法，如何只要求使用merge或者join可不可以完成任务四和任务五呢？
答 ： 因为Join和merge都是在水平方向上合并数据，所以需要append在垂直方向合并。
concat由于可以调整合并方向axis=0/1，但是不能根据指定条件合并。
